```python
import sys
sys.path.append('../')

import AGNCatalogToolbox as act
import numpy as np
from matplotlib import pyplot as plt

z = 2
```

```python
stellar_mass = np.linspace(9, 13)
black_hole_mass = act.stellar_mass_to_black_hole_mass(stellar_mass, scatter = None)

Geo = act.to_duty_cycle("Geo", stellar_mass, black_hole_mass, z=z, data_path="../Data/DutyCycles/")
Sch = act.to_duty_cycle("Schulze", stellar_mass, black_hole_mass, z=z, data_path="../Data/DutyCycles/")
Man = act.to_duty_cycle("Man16", stellar_mass, black_hole_mass, z=z, data_path="../Data/DutyCycles/")
```

```python
plt.figure()
plt.plot(stellar_mass, Geo, label = 'Geo')
plt.plot(stellar_mass, Sch, label = 'Sch')
plt.plot(stellar_mass, Man, label = 'Man')
plt.xlabel("Stellar Mass")
plt.ylabel("Duty Cycle")
plt.title("z = {}".format(z) )
plt.legend()
plt.savefig("DC_comparisonz{}.png".format(z))
plt.show()
```

```python
plt.figure()
plt.plot(black_hole_mass, Geo)
plt.plot(black_hole_mass, Sch)
plt.show()
```
